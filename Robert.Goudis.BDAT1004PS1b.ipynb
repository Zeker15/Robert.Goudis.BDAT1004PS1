Question 6. Python

```python
ay = 'ay'
way = 'way'
consonant = ('B','C','D','F','G','H','J','K','L','M','N','P','Q','R','S','T','V','W','X','Y','Z')
vowel = ('A','E','I','O','U')
pig_latin_string =''
user_sentence = input('Enter word to translate to Pig Latin:')
words = user_sentence.split()

for user_word in words:
    print(user_word)
    first_letter = user_word[0]
    first_letter = str(first_letter)
    first_letter = first_letter.upper()
    
if first_letter in consonant:
    length_of_word = len(user_word)
    remove_first_letter = user_word[1:length_of_word]
    pig_latin=remove_first_letter+first_letter+ay
    pig_latin_string=pig_latin_string+''+pig_latin
elif first_letter in vowel:
    pig_latin=user_word+way
    pig_latin_string=pig_latin+''+pig_latin
else:
    print('?')
    
print(pig_latin_string)

//https://stackoverflow.com/questions/54079026/pig-latin-program-in-python
```

```python
def bldcount('bloodtype1.txt'):

    infile = open(bloodtype1.txt, 'r')
    content = infile.read()
    content=content.split()
    bloodtype = ['A','B','AB','O',"OO"]
    for i in bloodtype:
        if content.count(i)==0:
            print('There are no patients with bloodtype',i+'.')
        elif content.count(i)==1:
            print('There is one patient with bloodtype',i+'.')
        else:
            print('There are',content.count(i),'patients that have bloodtype',i+'.')

bldcount('bloodtype1.txt')

```
